Nome: Grande Banco

Vers√£o do projeto: v1

Tr√™s opera√ß√µes:
- Dep√≥sito, Saque e Extrato

Regras:
- Somente valores positivos devem ser depositados na conta
- Limite de saques di√°rios = 3
- Limite por saque = R$ 500,00

```python
saldo = 700
saques = 0
depositos = 0

qtde_saques = 0
qtde_depositos = 0
limite_por_saque = 500
limite_saques = 3
operacao_contador = 1
extrato_historico = {}

operacoes_disponiveis = (1, 2, 3)

print("Seja bem-vindo ao Grande Banco!\nO maior banco do mundo a seu dispor. üèõ\n")
print("Qual opera√ß√£o voc√™ deseja realizar hoje?")

while True:
  acao = int(input('''
[1] Saque
[2] Dep√≥sito
[3] Extrato
'''))
  print()

  for operacao in operacoes_disponiveis:
    numero_operacao = f"Opera√ß√£o {operacao_contador}"
    if acao == operacoes_disponiveis[0]:
        valor = float(input("Qual valor voc√™ deseja sacar?\nAten√ß√£o: o limite por saque √© de R$ 500,00.\n"))
        print()
        if valor > saldo:
          print(f"Saldo insuficiente!\nSeu saldo atual √© de R$ {saldo:.2f}")
        elif valor > limite_por_saque:
          print("O valor m√°ximo por saque √© R$ 500,00\nTente novamente um valor menor ou outra opera√ß√£o na conta.\n")
        elif qtde_saques >= limite_saques:
          print("Voc√™ atingiu o limite de saques di√°rios!\n")
        else:
          saldo -= valor
          saques += valor
          qtde_saques += 1
          print(f"Saque de R$ {valor:.2f} realizado com sucesso!\n")
          extrato_historico[numero_operacao] = ["Saque", valor]
        break

    elif acao == operacoes_disponiveis[1]:
        valor = float(input("Qual valor voc√™ deseja depositar?\n"))
        if valor > 0:
          saldo += valor
          depositos += valor
          print(f"Dep√≥sito de R$ {valor:.2f} realizado com sucesso!\n")
          extrato_historico[numero_operacao] = ["Dep√≥sito", valor]
          qtde_depositos += 1
        else:
          print("\nValor inv√°lido para dep√≥sito!\nTente novamente.\n")
        break

    elif acao == operacoes_disponiveis[2]:
      if extrato_historico == {}:
        print("N√£o foram realizadas movimenta√ß√µes!\n")
        break

      else:
        print(f'''Extrato

Saldo atual: R$ {saldo:.2f}

Saques realizados: {qtde_saques} no total de R$ {saques:.2f}
Dep√≥sitos realizados: {qtde_depositos} no total de R$ {depositos:.2f}

Segue o hist√≥rico completo de movimenta√ß√µes:
        ''')
        for chave, valor in extrato_historico.items():
          print(f"{valor[0]} de R$ {valor[1]:.2f}")
        print()
        break

    elif acao != operacao:
        print("Opera√ß√£o inv√°lida!\nTente novamente.\n")
        break

  mais_acao = input("Deseja realizar mais alguma opera√ß√£o?\nDigite S para sim e N para N√£o\n").upper()
  operacao_contador += 1
  if mais_acao == "N":
    print("\nObrigado por utilizar nossos servi√ßos!\n")
    break
```

Tr√™s opera√ß√µes:

Agora √© permitido criar usu√°rio e criar conta corrente vinculada ao usu√°rio.

Regras:
- Usu√°rio cont√©m: nome, data de nascimento, CPF e endere√ßo.
- O endere√ßo √© composto por: logradouro, n√∫mero, bairro, cidade/Sigla do Estado.
- Deve ser armazenado somente os n√∫meros do CPF
- N√£o pode existir dois CPFs iguais
- Conta cont√©m: Ag√™ncia, n√∫mero da conta e usu√°rio vinculado
- O n√∫mero da conta √© sequencial (iniciando em 1)
- O n√∫mero da ag√™ncia √© fixo (0001)
- Um usu√°rio pode ter mais de uma conta
- O aplicativo s√≥ pode ser iniciado se uma conta cadastrada for informada.

```python
from datetime import date

usuarios = {
   "55944435620": ["Leandro", "07/02/2002", ["Avenida Bandeirantes", "457", "Miramar", "Sorocaba", "SP"], [1, 4]],
   "88744426532": ["Sandra", "12/04/1987", ["Rua Chic√≥ria", "356", "Vila Mariana", "S√£o Paulo", "SP"], [2]],
   "77966652420": ["Reinaldo", "05/04/1996", ["Rua Augusta", "1126", "Avenida Paulista", "S√£o Paulo", "SP"], [3, 5]]
}

contas = {
    1: ["0001", "55944435620", 750, {}],
    2: ["0001", "88744426532", 9005, {}],
    3: ["0001", "77966652420", 502.78, {}],
    4: ["0001", "55944435620", 0, {}],
    5: ["0001", "77966652420", 45500.56, {}]
}

def criar_usuario():
  nome = input("\nQual o seu nome?\n")
  cpf = str(input("\nDigite o seu CPF:\n"))
  cpf = cpf.replace(".", "").replace("-", "")
  if len(cpf) != 11:
    print("CPF inv√°lido!")
    return "CPF inv√°lido"
  elif cpf in usuarios.keys():
    print("CPF j√° cadastrado!")
    return "CPF j√° cadastrado"

  else:
    print("\n√â a sua primeira vez sua com a gente! Vamos continuar criando seu usu√°rio.\n")
    data_nascimento = input("Qual a sua data de nascimento?\nPreencha seguindo o padr√£o 'Dia/M√™s/Ano'. Exemplo: 16/04/2000\n")

    dia, mes, ano = data_nascimento.split('/')
    if date.today().year - int(ano) < 18:
        print(f"\n{nome}, n√£o √© poss√≠vel criar uma conta para voc√™ no momento. Voc√™ precisa ter pelo menos 18 anos de idade para criar uma conta.\n\nPe√ßa para o seu respons√°vel intermediar o processo para voc√™.")
        return

    else:
        print("\nVamos agora completar o seu endere√ßo!\nMe conta, qual √© o seu...")
        logradouro = input("Logradouro (Avenida, Rua...): ")
        numero = input("N√∫mero da casa: ")
        bairro = input("Bairro: ")
        cidade = input("Cidade: ")
        estado = input("Estado: ")
        endereco = [logradouro, numero, bairro, cidade, estado]

        usuarios[cpf] = [nome, data_nascimento, endereco, []]
        print("\nUsu√°rio criado com sucesso!\n")

def continuar_operacao():
    continuar_operacao = input("\nPor favor, digite 1 para tentar novamente ou 0 para encerrar essa opera√ß√£o.\n")
    if continuar_operacao != "0":
      novo_usuario = criar_usuario()
    else:
      print("\nObrigado por utilizar nossos servi√ßos!\n")
      return "0"

print("Seja bem-vindo ao Grande Banco!\nO maior banco do mundo a seu dispor. üèõ\n")
usuario_cadastrado = input("Voc√™ j√° possui uma conta?\nDigite S para sim e N para N√£o\n").upper()

if usuario_cadastrado == "N":
  print("\nVamos criar uma conta para voc√™!")
  novo_usuario = criar_usuario()

  while True:
    if novo_usuario == "CPF inv√°lido":
      operacao = continuar_operacao()
      if operacao == "0":
        break
    elif novo_usuario == "CPF j√° cadastrado":
      operacao = continuar_operacao()
      if operacao == "0":
        break
    else:
      break
```
